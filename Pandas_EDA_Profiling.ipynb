```python
import pandas as pd
from pandas_profiling import ProfileReport
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
profile = ProfileReport(iris, title = "Iris Data Profile")
```

```python
#for google colab
#profile.to_notebook_iframe()
#for jupyter notebook
profile.to_widgets()
```

```python
profile.to_file("relatorio.html")
```
