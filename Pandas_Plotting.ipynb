```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
daterange = pd.period_range('1/1/1950', freq = '1d', periods = 50)
date_df = pd.DataFrame(data = daterange, columns = ['day'])
date_df['value1'] = np.random.randint(45, 65, size = (len(date_df)))
date_df['value2'] = np.random.randint(25, 35, size = (len(date_df)))
date_df.head(3)
```

```python
ax = date_df.plot(x='day', y=['value1', 'value2'])
```

```python
date_df.plot.area(stacked = True)
```

```python
date_df.plot.area(stacked = False)
```

```python
iris = pd.read_csv('Iris.csv')
```

```python
iris.hist()
```

```python
iris.head(3)
```

```python
iris[['SepalWidthCm', 'SepalLengthCm']].plot.hist(alpha = 0.5)
```

```python
iris['Color'] = iris.apply(lambda x: 'red' if x['Species'] == 'Iris-setosa' else ('green' if x['Species'] == 'Iris-virginica' else 'blue'), axis=1)
iris.plot.scatter(x='SepalWidthCm', y='SepalLengthCm', color = iris['Color'])
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(iris, figsize = (15, 9),)
```
